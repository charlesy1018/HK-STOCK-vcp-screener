```python
url='https://raw.githubusercontent.com/charlesy1018/HKSTOCK-market-daily-check/main/HangSeng_Index_StockList.csv'
HSI_stock_ticker=pd.read_csv(url)
print(HSI_stock_ticker)
```

```python
HSI_symbol = HSI_stock_ticker.Symbol.to_list()
print(HSI_symbol)
```

```python
url2= "https://raw.githubusercontent.com/charlesy1018/HKSTOCK-market-daily-check/main/Hang%20Seng%20MidCap-StockList.csv"
midcap_stock_ticker=pd.read_csv(url2)
print(midcap_stock_ticker)
```

```python
midcap_symbol= midcap_stock_ticker.Symbol.to_list()
print(midcap_symbol)
```

```python
#merging the two list together
for symbol in midcap_symbol:
    HSI_symbol.append(symbol)
    
HSI_symbol = list(dict.fromkeys(HSI_symbol))
print(HSI_symbol)
```

```python
len(HSI_symbol)
```

```python
def getdata(tickers):
    data=[]
    for ticker in tickers:
        data.append(yf.download(ticker,period='5y').reset_index())
    return data
```

```python
HK = getdata(HSI_symbol)
```

```python
HK
```

```python
#Assigning symbol into each stock list
for i in range(len(HK)):
    HK[i]['Symbol']= HSI_symbol[i]
print(HK)
```

```python
def createengine(name):
  engine = sqlalchemy.create_engine('sqlite:///'+name)
  return engine
```

```python
HKengine = createengine('HK')
```

```python
def TOSQL(frames, symbols,engine):
  for frame,symbol in zip(frames,symbols):
      frame.to_sql(symbol,engine,index=False)
  print('Successfully imported data')

#frames 同 symbols 係2個唔同list, combine 2個唔同list and add to sql server
```

```python
TOSQL(HK,HSI_symbol,HKengine)
```
